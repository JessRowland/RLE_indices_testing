```python
# Load pickled inputs
p = open('_pickle/reefmap','r'); reefmap = pickle.load(p); p.close()
p = open('_pickle/params','r'); params = pickle.load(p); p.close()
p = open('_pickle/transitionCs','r'); transitionCs=pickle.load(p); p.close()
p = open('_pickle/transitionF','r'); transitionF = pickle.load(p); p.close()
p = open('_pickle/transitionU','r'); transitionU = pickle.load(p); p.close()
p = open('_pickle/transition','r'); transition = pickle.load(p); p.close()
p = open('_pickle/forcing_nut','r'); forcing_nut = pickle.load(p); p.close()
p = open('_pickle/forcing_sed','r'); forcing_sed = pickle.load(p); p.close()
p = open('_pickle/forcing_hurr','r'); forcing_hurr=pickle.load(p); p.close()
p = open('_pickle/forcing_cm','r'); forcing_cm = pickle.load(p); p.close()
p = open('_pickle/forcing_f','r'); forcing_f = pickle.load(p); p.close()
p = open('_pickle/forcing_df','r'); forcing_df = pickle.load(p); p.close()
```

```python
 # Lists and dictionaries to catch model results
defectruns = []
mcavs = {}; mcCIs = {}; mcavsSR = {}; mcCIsSR = {}; cellavs = {}
hlog_cat = []; hlog_sr = []; cmlog_sr = []
# Put the variables within var into the dictionaries
for var in variables:
    mcavs[var] = []; mcCIs[var] = []; mcavsSR[var] = []; mcCIsSR[var] = []; 
    cellavs[var] = []
coralmaps = [[]  for run in range(params.runs)]
```

```python
coralmaps
```

```python
run = 1
model = Simulation(reefmap,params)
transitionCb = Transition()
transitionCb.makeTMR(transition.pctotal,params.paramrec['dlarv_Cb'],run)
model.model(reefmap,params,variables,transition,transitionCb,transitionCs,
            transitionF,transitionU,forcing_nut,forcing_sed,forcing_hurr,
            forcing_cm,forcing_f,forcing_df,run)
if model.errors > 0: defectruns.append(run)
if model.errors == 0:
    model.averages(reefmap,params,variables,100,1)
    model.averagesSR(reefmap,params,variables,100,1)
    for var in variables:
        mcavs[var].append(model.avs[var])
        mcCIs[var].append(model.CIs[var])
        mcavsSR[var].append(model.avsSR[var])
        mcCIsSR[var].append(model.CIsSR[var])
        cellavs[var].append(model.modelout[var][-1])
    hlog_cat.append(model.hlog_cat); hlog_sr.append(model.hlog_sr)
    cmlog_sr.append(model.cmlog_sr)
    coralmaps[run] = model.modelout['C']
```
